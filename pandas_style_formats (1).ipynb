```python
#importing library and dataset over seaborn
import seaborn as sns
df=sns.load_dataset("titanic")
df.head()
```

```python
#get top 10 numeric features
df_numeric=df._get_numeric_data().head(10)
df_numeric
```

```python
#highlight max values in each columns
def highlight_max(s):
    is_max = s == s.max()
    return ['background-color: yellow' if v else '' for v in is_max]

df_numeric.style.apply(highlight_max)
```

```python
df.head(25).style.bar(
    color="lightgreen")
```

```python
#change background color and color of text
df_numeric.style.set_properties(**{'background-color': 'black',
                            'color': 'lawngreen',
                            'border-color': 'white'})
```

```python
#Bar Charts Styling
df_numeric.style.bar(subset=['age', 'fare'], color='#d65f5f')
```

```python
#styling just desired column>fare
df_numeric.style.bar(subset=['fare'], color='green')
```

```python
# Compress the color range
(df.loc[:9]
    .style
    .background_gradient(cmap='viridis', low=.5, high=0)
    .highlight_null('red'))
```

```python
import seaborn as sns
df=sns.load_dataset("tips")
df.head()
```

```python
# Compress the color range
(df.loc[:9]
    .style
    .background_gradient(cmap='viridis', low=.5, high=0)
    .highlight_null('red'))
```

```python
df.head(15).style.set_properties(**{'background-color': 'black',
                           'color': 'red'})

```

```python
# giving border to dataframe
df.head(10).style.set_table_styles([{'selector' : '',
                            'props' : [('border',
                                        '20px solid blue')]}])
```

```python
# generating pairwise correlation
corr = df.corr()
  
# Displaying dataframe as an heatmap 
# with diverging colourmap as coolwarm
corr.style.background_gradient(cmap ='coolwarm')
```

```python
# Declaring the cm variable by the
# color palette from seaborn
cm = sns.light_palette("green", as_cmap=True)
 
# Visualizing the DataFrame with set precision
print("\nModified Stlying DataFrame:")
df.style.background_gradient(cmap=cm).set_precision(2)
```

```python
# Python program to generate a heatmap
# which displays the value in each cell 
# corresponding to the given dataframe 
  
# import required libraries
import pandas as pd
  
# defining index for the dataframe
idx = ['1', '2', '3', '4']
  
# defining columns for the dataframe
cols = list('ABCD')
  
# entering values in the index and columns  
# and converting them into a panda dataframe
df = pd.DataFrame([[10, 20, 30, 40], [50, 30, 8, 15],
                   [25, 14, 41, 8], [7, 14, 21, 28]],
                   columns = cols, index = idx)
  
# displaying dataframe as an heatmap 
# with diverging colourmap as virdis
df.style.background_gradient(cmap ='viridis')\
        .set_properties(**{'font-size': '20px'})
```
